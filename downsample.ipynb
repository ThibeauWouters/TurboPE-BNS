# Downsample

```python
import numpy as np
import matplotlib.pyplot as plt
import corner
import os

downsample_factor = 20
labels = ["TF2", "NRTv2"]
target_dir = "/home/thibeau.wouters/datasets/TurboPE_BNS_injection/"
```

```python
# Iterate over the 2 labels for the 2 waveforms
for label in labels:
    # Fetch the correct directory
    dir = f"./injections/outdir_{label}/"
    # Iterate over each injection
    injection_dirs = os.listdir(dir)
    
    for injection in injection_dirs:
        # Construct the paths
        path = os.path.join(dir, injection_dirs[0], "results_production.npz")
        out_path = os.path.join(target_dir, f"injections_{label}", f"{injection}_downsampled.npz")
        
        # Load the data
        data = np.load(path)
        chains = data['chains']
        log_prob = data['log_prob']
        
        chains = chains.reshape((-1, chains.shape[-1]))
        log_prob = log_prob.reshape((-1,))
        
        # Downsample
        chains = chains[::downsample_factor]
        log_prob = log_prob[::downsample_factor]
        
        # Save to new file
        print(f"Saving to {out_path}")
        np.savez(out_path, chains=chains, log_prob=log_prob)
```

# Load again as test

```python
# Load most recently saved data
data = np.load(out_path)
chains = data['chains']
log_prob = data['log_prob']

print("np.shape(chains)")
print(np.shape(chains))

# Test with corner if OK

fig = corner.corner(chains)
plt.show()
plt.close()
```
